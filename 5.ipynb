$$z = 1.7∗exp[-\{\frac{(x − 3)^{2}}{10} + \frac{(y-3)^{2}}{10}\}]+exp[-\{\frac{(x + 5)^{2}}{8} + \frac{(y+5)^{2}}{8}\}]+2*exp[-\{\frac{x^{2}}{4}+\frac{y^{2}}{5}\}]+1.5*exp[-\{\frac{(x-4)^{2}}{18} + \frac{(y+4)^{2}}{16}\}] + 1.2*exp[-\{\frac{(x+4)^{2}}{18} + \frac{(y-4)^{2}}{16}\}]$$<br><br>

## 1. Gradient Ascent
<br><br><b>Find the maxima $z^{*} = f(x^{*},y^{*})$ using Gradient Ascent. Experiment with multiple initial
values $(x^{(0)}, y^{(0)}) \in [−10, 10] × [−10, 10]$ and different number of iterations. Visualize the
contour plot of $f(x, y)$ and show the trajectories of the gradient ascent solution iterations.
Report the best solutions $x^{*},y^{*},z^{*}$.

<br><b>Initialised $x_{0} = 5$, $y_{0} = 5$, $\alpha = 0.05$, $n = 1000$

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 0.05
n = 1000
x_iter = np.ones(n+1)
y_iter = np.ones(n+1)
z_iter = np.ones(n)
x_iter[0] = 5
y_iter[0] = 5

for i in range(n):
    t1i = 1.7 * np.exp(-1/10*(np.square(x_iter[i]-3)+ np.square(y_iter[i]-3)))
    t2i = np.exp(-1/8*(np.square(x_iter[i]+5) + np.square(y_iter[i]+5)))
    t3i = 2 * np.exp(-1*(np.square(x_iter[i])/4 + np.square(y_iter[i])/5))
    t4i = 1.5 * np.exp(-1 * (np.square(x_iter[i]-4)/18 + np.square(y_iter[i]+4)/16))
    t5i = 1.2 * np.exp(-1*(np.square(x_iter[i]+4)/18 + np.square(y_iter[i]-4)/16))
    z_iter[i] = t1i  + t2i + t3i + t4i + t5i
    
    dt1xi = t1i * (-2/10*(x_iter[i]-3))
    dt2xi = t2i * (-2/8*(x_iter[i]+5)) 
    dt3xi = t3i * (-2/4*x_iter[i])
    dt4xi = t4i * (-2/18*(x_iter[i]-4))
    dt5xi = t5i * (-2/18*(x_iter[i]+4))
    
    dt1yi = t1i * (-2/10*(y_iter[i]-3))
    dt2yi = t2i * (-2/8*(y_iter[i]+5))
    dt3yi = t3i * (-2/5*y_iter[i])
    dt4yi = t4i * (-2/16*(y_iter[i]+4))
    dt5yi = t5i * (-2/16*(y_iter[i]-4))
    
    dx = dt1xi +dt2xi +dt3xi +dt4xi +dt5xi
    dy = dt1yi +dt2yi +dt3yi +dt4yi +dt5yi
    
    x_iter[i+1] = x_iter[i] + alpha * dx
    y_iter[i+1] = y_iter[i] + alpha * dy
        
x_iter = np.array(x_iter)[:-1]
y_iter = np.array(y_iter)[:-1]


x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)  
z = np.zeros((np.shape(x)[0],np.shape(y)[0]))
zmax = -150000
for i in range(np.shape(x)[0]):
    for j in range(np.shape(y)[0]):
        t1ij = 1.7 * np.exp(-1/10*(np.square(x[i]-3)+ np.square(y[j]-3)))
        t2ij = np.exp(-1/8*(np.square(x[i]+5) + np.square(y[j]+5)))
        t3ij = 2 * np.exp(-1*(np.square(x[i])/4 + np.square(y[j])/5))
        t4ij = 1.5 * np.exp(-1 * (np.square(x[i]-4)/18 + np.square(y[j]+4)/16))
        t5ij = 1.2 * np.exp(-1*(np.square(x[i]+4)/18 + np.square(y[j]-4)/16))
        z[i][j] = t1ij  + t2ij + t3ij + t4ij + t5ij
        if(z[i,j] > zmax):
            zmax = z[i, j]
            x_max, y_max = x[i], y[j]
        
        
        
X, Y = np.meshgrid(x, y)
fig, ax3 = plt.subplots(1, 1)
ax3.contour(X, Y, z)
plt.plot(x_iter, y_iter, marker=".", markersize=0.1, color = 'black')

plt.show()

print(f'Global maxima of function is near  f({x_max},{y_max}) = {zmax}')
print(f'Predicted values by gradient ascent are f({x_iter[-1]},{y_iter[-1]}) = {z_iter[-1]}')
```

**Result**: *As we can see the gradient ascent has reached the global maxima*<br><br><br>

## <br><br><br>2. Stochastic Search

<br><br><b>Find the maxima $z^{*} = f(x^{*},y^{*})$ using Stochastic Search. The solution search space is given
by the search space bounds $X_{min} = [x_{min}, y_{min}]^{T} = [−10, −10]^{T}$ and $X_{max} = [x_{max}, y_{max}]^{T} = [−10,−10]^{T}$. Write the following function in Python.<br>
<br>$$[bestX, bestY, maxF] = stochasticSearch(X_{min}, X_{max}, popSize, nbhSize, maxItr)$$<br>
Here,$z^{*} = maxF = f(bestX = x^{*}, bestY = y^{*})$ is the best solution found by Stochastic Search with a Solution Population Size of popSize and maxItr Iterations (or Generations). During pure exploitation, the children solutions of a parent px are generated in a hypersphere of radius nbhSize centered at $px$. Display the scatter plot of the solutions in each
iteration on the contour plot of $f(x, y)$ to visualize the trajectories of the solutions in the population. Experiment with different values of popSize, nbhSize and maxItr and report the best solution.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def stochasticSearch(Xmin, Xmax, popSize, nbhSize, maxItr):
    xmin=Xmin[0]
    ymin=Xmin[1]
    xmax=Xmax[0]
    ymax=Xmax[1]
    X=np.zeros((2,popSize,maxItr))
    Y= np.zeros((maxItr+1,popSize))
    U=np.zeros((maxItr,popSize))
    P=np.zeros((maxItr,popSize))
    C=np.zeros((1,popSize))

    #Intialising the population
    for i in range(popSize):
        alpha1=np.random.uniform(0,1)
        alpha2=np.random.uniform(0,1)
        X[0,i,0]=(1-alpha1)*xmin+(alpha1*xmax)
        X[1,i,0]=(1-alpha2)*ymin+(alpha2*ymax)
        t1i = 1.7 * np.exp(-1/10*(np.square(X[0,i,0]-3)+ np.square(X[1,i,0]-3)))
        t2i = np.exp(-1/8*(np.square(X[0,i,0]+5) + np.square(X[1,i,0]+5)))
        t3i = 2 * np.exp(-1*(np.square(X[0,i,0])/4 + np.square(X[1,i,0])/5))
        t4i = 1.5 * np.exp(-1 * (np.square(X[0,i,0]-4)/18 + np.square(X[1,i,0]+4)/16))
        t5i = 1.2 * np.exp(-1*(np.square(X[0,i,0]+4)/18 + np.square(X[1,i,0]-4)/16))
        Y[0,i] = t1i  + t2i + t3i + t4i + t5i

    for i in range(maxItr):
        Si=X[:,:,i]
        #random walk
        Sxrw=[]
        Syrw=[]
        for j in range(popSize):
            U[i,j]=Y[i,j]-np.min(Y[i,:])
        for k in range(popSize):
            P[i,k]=U[i,k]/(np.sum(U[i,:]))
        for j in range(popSize):
            c=round(popSize*P[i,j])
            x=np.zeros((2,c))
            for k in range(c):
                v=np.random.uniform(-1,1,2)
                lmd=nbhSize*random.random()
                x[0,k]=X[0,k,i]+(lmd*(v[0]/np.linalg.norm(v)))
                x[1,k]=X[1,k,i]+(lmd*(v[1]/np.linalg.norm(v)))
                Sxrw.append(x[0,k])
                Syrw.append(x[1,k])
        Srw=np.array([Sxrw,Syrw])
    
        #random linear combination
        Sxrlc=[]
        Syrlc=[]
        for j in range(popSize-1):
            for k in range(j+1,popSize):
                alpha=np.random.uniform(0,1)
                x=(1-alpha)*X[0,j,i]+(alpha*X[0,k,i])
                y=(1-alpha)*X[1,j,i]+(alpha*X[1,k,i])
                Sxrlc.append(x)
                Syrlc.append(y)
        Srlc=np.array([Sxrlc,Syrlc])
    
        #random reintialisation
        Sxrri=[]
        Syrri=[]
        for j in range(popSize):
            alpha1=np.random.uniform(0,1)
            alpha2=np.random.uniform(0,1)
            x=(1-alpha1)*xmin+(alpha1*xmax)
            y=(1-alpha2)*ymin+(alpha2*ymax)
            Sxrri.append(x)
            Syrri.append(y) 
        Srri=np.array([Sxrri,Syrri])
    
        Su=np.unique(np.column_stack((Si,Srw,Srlc,Srri)),axis=1)
        y=np.zeros((Su.shape[1]))
        for j in range(Su.shape[1]):
            t1i = 1.7 * np.exp(-1/10*(np.square(Su[0,j]-3)+ np.square(Su[1,j]-3)))
            t2i = np.exp(-1/8*(np.square(Su[0,j]+5) + np.square(Su[1,j]+5)))
            t3i = 2 * np.exp(-1*(np.square(Su[0,j])/4 + np.square(Su[1,j])/5))
            t4i = 1.5 * np.exp(-1 * (np.square(Su[0,j]-4)/18 + np.square(Su[1,j]+4)/16))
            t5i = 1.2 * np.exp(-1*(np.square(Su[0,j]+4)/18 + np.square(Su[1,j]-4)/16))
            y[j] = t1i  + t2i + t3i + t4i + t5i
        ind=np.argpartition(y, -popSize)[-popSize:]
        Sb=Su[:,ind]
        Y[i+1,:]=y[ind]
        x = np.linspace(-10,10,100)
        y = np.linspace(-10,10,100) 
        z = np.zeros((np.shape(x)[0],np.shape(y)[0]))
        for i in range(np.shape(x)[0]):
            for j in range(np.shape(y)[0]):
                t1ij = 1.7 * np.exp(-1/10*(np.square(x[i]-3)+ np.square(y[j]-3)))
                t2ij = np.exp(-1/8*(np.square(x[i]+5) + np.square(y[j]+5)))
                t3ij = 2 * np.exp(-1*(np.square(x[i])/4 + np.square(y[j])/5))
                t4ij = 1.5 * np.exp(-1 * (np.square(x[i]-4)/18 + np.square(y[j]+4)/16))
                t5ij = 1.2 * np.exp(-1*(np.square(x[i]+4)/18 + np.square(y[j]-4)/16))
                z[i][j] = t1ij  + t2ij + t3ij + t4ij + t5ij

        X1, Y1 = np.meshgrid(x, y)
        fig, ax3 = plt.subplots(1, 1)
        ax3.contour(X1, Y1, z)
        plt.scatter(Sb[0,:],Sb[1,:])
        plt.show()



    ind1=np.argmax(Y[maxItr,:])
    
    return Sb[0,ind1],Sb[1,ind1],np.max(Y[maxItr,:])
    
   
        
```

$popSize=10$, $nbhSize=1$, $maxItr=10$

```python
Xmin=[-10,-10]
Xmax=[10,10]
popSize=10
maxItr=10
nbhSize=1
xmax,ymax,Fmax=stochasticSearch(Xmin, Xmax, popSize, nbhSize, maxItr)
print(f'Predicted values by stochastic Search are f({xmax},{ymax}) = {Fmax}')
```
