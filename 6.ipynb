<br><b>Q1. Consider the city problem described by the att48.tsp dataset. Consider a Tour starting from city and ending at city . Search for an appropriate travel itinerary involving visits to all the remaining cities in a certain sequence while minimizing the total tour length. Solve this problem using stochastic search. Choose appropriate stochastic search parameters.<br><br>
(a) Plot the algorithm progress i.e. best distance value in each iteration.<br><br> 
(b)Plot the best tour (path connecting city sequence) obtained after each K (user choice for plotting) iterations.

```python
import numpy as np
import random
import copy
import matplotlib.pyplot as plt

### Reading the text file and storing the distances in a 2 dimensional array

Y = [int(x) for x in np.linspace(1,48,48)]
Read = []
with open('att48_d.txt', encoding='utf8') as f:
    for line in f:
        Read.append([int(x) for x in line.split()])
        
M=np.array(Read)

### coordinates of the points
coordinates = []
with open("att48_xy.txt", encoding='utf8') as f:
    for points in f:
        coordinates.append([int(x) for x in points.split()])
        
###function to find the path distance
def fitness(Newlist):
    FitnessValue = []
    for Paths in Newlist:
        fitness = 0
        for i in range(len(Paths)):
            if i == len(Paths) -1 :
                fitness += M[Y.index(Paths[i])][0]
            else:
                fitness += M[Y.index(Paths[i])][Y.index(Paths[i+1])]
                
        FitnessValue.append(-fitness)
        
    return FitnessValue

###function to plot the path
def plotpath(parentslist):
    co_num = np.array(coordinates)
    co_num = co_num[np.array(parentslist)-1]
    y = co_num[:,1]
    x = co_num[:,0]
    fig = plt.figure(figsize=(10,10))
    plt.plot(co_num[:,0],co_num[:,1],"r")
    plt.quiver(x[:-1], y[:-1], x[1:]-x[:-1], y[1:]-y[:-1], scale_units='xy', angles='xy', scale=1)
    plt.scatter(co_num[:,0],co_num[:,1])
    for i, label in enumerate(parentslist):
        plt.annotate(label, (co_num[i]))
        
    plt.show()
    parentslist

##function to generate the population
def population(list1,n):
    x = []
    U = list1.copy()
    if 1 in U:
        U.remove(1)
    for s in range(n):
        random.shuffle(U)
        t = [1] + U + [1]
        x.append(t)
        
    return x
        
#Intialising the population
popSize=100
maxItr=1000
St=population(Y,popSize)
St=np.array(St)
Yt=fitness(St)

for i in range(maxItr):
    #random swap for children generation
    P=[]
    Srw=[]
    u=[]
    for l in range(popSize):
        u.append(Yt[l]-min(Yt))
    for k in range(popSize):
        P.append(u[k]/(np.sum(u)))
    for j in range(popSize):
        c=round(popSize*P[j])
        
        for k in range(c):
            r=random.sample(range(1, 47), 4)
            S=St[j].copy()
            S[r[0]],S[r[1]] = S[r[1]],S[r[0]]
            S[r[2]],S[r[3]] = S[r[3]],S[r[2]]
            Srw.append(S)
    Srw=np.array(Srw)
    Su=np.row_stack((St,Srw))
    Yt=fitness(np.array(Su).tolist())
    Su=np.array(Su)
    Yt=np.array(Yt)
    
    ind=np.argpartition(Yt, -popSize)[-popSize:]
    
    St=Su[ind,:]
    Yt=Yt[ind]
    ind1=np.argmax(Yt)
    print("Best Distance at iteration: ",i+1,"is ",-Yt[ind1])
    if (i+1)%100 == 0:
        plotpath(St[ind1,:])
print("\nThe best path obtained: ",St[ind1,:])
print("\nThe Best Distance: ",-Yt[ind1])
```

<br><br><b>Q2. Maximization using PSO<br><br>
$$z = 1.7exp[-\{\frac{(x  3)^{2}}{10} + \frac{(y-3)^{2}}{10}\}]+exp[-\{\frac{(x + 5)^{2}}{8} + \frac{(y+5)^{2}}{8}\}]+2*exp[-\{\frac{x^{2}}{4}+\frac{y^{2}}{5}\}]+1.5*exp[-\{\frac{(x-4)^{2}}{18} + \frac{(y+4)^{2}}{16}\}] + 1.2*exp[-\{\frac{(x+4)^{2}}{18} + \frac{(y-4)^{2}}{16}\}]$$<br><br>

<b>Find the maxima $ = (  ,  )$ using PSO with a solution population size of  and  Iterations. The solution search space is given by the search space bounds $_{min} = [10, 10]^{T}$ and $X_{max} = [10,10]^{T}$. Write the following function in Python.<br><br>
$$[bestX,bestY,maxF]=PSO(X_{min},X_{max},popSize,maxItr,psoParams)$$<br><br>
Display the scatter plot of the solutions in each iteration on the contour plot of $(, )$ to visualize the trajectories of the solutions in the population. Experiment with different values of PSO parameters $(), $ and $$ and report the best solution.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def PSO(Xmin, Xmax, popSize, maxItr, omg, phip, phig):
    xmin=Xmin[0]
    ymin=Xmin[1]
    xmax=Xmax[0]
    ymax=Xmax[1]
    X=np.zeros((2,popSize))
    v=np.zeros((2,popSize))
    p=np.zeros((2,popSize))
    y=np.zeros((1,popSize))
    pb=np.zeros((1,popSize))
    g=np.zeros((2,1))

    #Intialising the population
    for i in range(popSize):
        alpha1=np.random.uniform(0,1)
        alpha2=np.random.uniform(0,1)
        X[0,i]=(1-alpha1)*xmin+(alpha1*xmax)
        X[1,i]=(1-alpha2)*ymin+(alpha2*ymax)
        p[:,i]=X[:,i]
        t1i = 1.7 * np.exp(-1/10*(np.square(X[0,i]-3)+ np.square(X[1,i]-3)))
        t2i = np.exp(-1/8*(np.square(X[0,i]+5) + np.square(X[1,i]+5)))
        t3i = 2 * np.exp(-1*(np.square(X[0,i])/4 + np.square(X[1,i])/5))
        t4i = 1.5 * np.exp(-1 * (np.square(X[0,i]-4)/18 + np.square(X[1,i]+4)/16))
        t5i = 1.2 * np.exp(-1*(np.square(X[0,i]+4)/18 + np.square(X[1,i]-4)/16))
        y[0,i] = t1i  + t2i + t3i + t4i + t5i
    gbk=np.argmax(y)
    gb=y[0,gbk]
    g[:,0]=X[:,gbk]

    for i in range(maxItr):
        beta=np.random.uniform(0,1,(2,1))
        gama=np.random.uniform(0,1,(2,1))
        v=(omg*v)+(phip*beta*(p-X))+(phig*gama*(g-X))
        X=X+v
        for j in range(popSize):
            t1i = 1.7 * np.exp(-1/10*(np.square(X[0,j]-3)+ np.square(X[1,j]-3)))
            t2i = np.exp(-1/8*(np.square(X[0,j]+5) + np.square(X[1,j]+5)))
            t3i = 2 * np.exp(-1*(np.square(X[0,j])/4 + np.square(X[1,j])/5))
            t4i = 1.5 * np.exp(-1 * (np.square(X[0,j]-4)/18 + np.square(X[1,j]+4)/16))
            t5i = 1.2 * np.exp(-1*(np.square(X[0,j]+4)/18 + np.square(X[1,j]-4)/16))
            y[0,j] = t1i  + t2i + t3i + t4i + t5i
        
            t1i = 1.7 * np.exp(-1/10*(np.square(p[0,j]-3)+ np.square(p[1,j]-3)))
            t2i = np.exp(-1/8*(np.square(p[0,j]+5) + np.square(p[1,j]+5)))
            t3i = 2 * np.exp(-1*(np.square(p[0,j])/4 + np.square(p[1,j])/5))
            t4i = 1.5 * np.exp(-1 * (np.square(p[0,j]-4)/18 + np.square(p[1,j]+4)/16))
            t5i = 1.2 * np.exp(-1*(np.square(p[0,j]+4)/18 + np.square(p[1,j]-4)/16))
            pb[0,j] = t1i  + t2i + t3i + t4i + t5i
        
            if y[0,j]>pb[0,j]:
                p[:,j]=X[:,j]
        
        k=np.argmax(pb)
    
        if pb[0,k]>gb:
            g[:,0]=p[:,k]
            gb=pb[0,k]    
            gbk=k
            
        x1 = np.linspace(-10,10,100)
        y1 = np.linspace(-10,10,100) 
        z = np.zeros((np.shape(x1)[0],np.shape(y1)[0]))
        for i in range(np.shape(x1)[0]):
            for j in range(np.shape(y1)[0]):
                t1ij = 1.7 * np.exp(-1/10*(np.square(x1[i]-3)+ np.square(y1[j]-3)))
                t2ij = np.exp(-1/8*(np.square(x1[i]+5) + np.square(y1[j]+5)))
                t3ij = 2 * np.exp(-1*(np.square(x1[i])/4 + np.square(y1[j])/5))
                t4ij = 1.5 * np.exp(-1 * (np.square(x1[i]-4)/18 + np.square(y1[j]+4)/16))
                t5ij = 1.2 * np.exp(-1*(np.square(x1[i]+4)/18 + np.square(y1[j]-4)/16))
                z[i][j] = t1ij  + t2ij + t3ij + t4ij + t5ij

        X1, Y1 = np.meshgrid(x1, y1)
        fig, ax3 = plt.subplots(1, 1)
        ax3.contour(X1, Y1, z)
        plt.scatter(X[0,:],X[1,:])
        plt.show()
        
    return X[0,gbk],X[1,gbk],gb
```

<br>$popSize=10,maxItr=10,\omega=0.3,\phi_{p}=0.25,\phi_{g}=0.4$

```python
Xmin=[-10,-10]
Xmax=[10,10]
popSize=10
maxItr=10
omg=0.3
phip=0.25
phig=0.4
xmax,ymax,Fmax=PSO(Xmin, Xmax, popSize, maxItr, omg, phip, phig)
print(f'Predicted values by PSO are f({xmax},{ymax}) = {Fmax}')
```
